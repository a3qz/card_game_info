```python
from bs4 import BeautifulSoup
import urllib
from lxml import html
import requests 
import string
import pprint
```

```python
prefix = "http://www.pagat.com/"
page = requests.get('http://www.pagat.com/alpha/')
tree = html.fromstring(page.content)
players = tree.xpath('//span[@class="buyer-name"]/text()')
unformatted_link = tree.xpath('//a[text()]/@href')
working_link = unformatted_link
new_link = []
for i in working_link:
    if '..'  in i:
        new_link.append(string.replace(i,'..', prefix))
for i in new_link:
    print i

alpha_games_list = new_link[7 : 1309]
```

```python
pages = []
i = 0
for games in alpha_games_list:
    print i
    f = urllib.urlopen(games)
    pages.append(BeautifulSoup(f))
    f.close()
    i += 1
```

```python
pages_text = []
for i in pages:
    pages_text.append(i.get_text().encode('ascii', 'ignore').decode('ascii'))
```

```python
print pages_text[0][0:80]
```

```python
count = 0
for i in pages_text:
    name_1 = "./summaries/"
    name_2 = str(count)
    name_full = name_1 + name_2
    output = open(name_full, 'write')
    output.write("#################################################")
    output.write(str(pages_text))
    output.close()
    count = count + 1
```
